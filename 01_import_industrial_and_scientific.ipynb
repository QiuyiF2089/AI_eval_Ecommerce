# Import Industrial_and_Scientific item metadata (Hugging Face datasets)

This notebook downloads the Industrial_and_Scientific **item metadata** dataset from Hugging Face `datasets` and stores it under `data/`.


```python
from pathlib import Path
import shutil

data_dir = Path('data')
if data_dir.exists():
    shutil.rmtree(data_dir)
data_dir.mkdir(parents=True, exist_ok=True)
```

```python
import datasets
from datasets import load_dataset

datasets.logging.set_verbosity_error()

dataset = load_dataset(
    'McAuley-Lab/Amazon-Reviews-2023',
    'raw_meta_Industrial_and_Scientific',
    split='full',
    trust_remote_code=True
)
dataset
```

```python
# Save locally under data/
out_dir = data_dir / 'raw_meta_Industrial_and_Scientific_hf'
dataset.save_to_disk(out_dir)
out_dir
```
